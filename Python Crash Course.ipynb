Rules:
* Variables start with an _ or a letter, never with a number
* Variables are case sensitive
* Variables must be assigned before being referenced

Conventions:
*  Variable names should be lowercase, with words separated by underscores as necessary to improve readability.

**Arithmetic Operations**<br>
We use arithmetic operators to do math.

````
sum = 2 + 3
difference = 2 - 3
product = 2 * 3
division = 2 / 3
integer_division = 2 // 3
modulo = 5 % 3

print(sum, difference, product, division, integer_division, modulo)
````

````
square = 3 ** 2
result = square + 3 * (4 - 5) / 3
print(square, result)
````

**If Statements** <br>
Again, indentation instead of brackets.

````
if 2 < 3:
  print('It Works!')
````

You can add multiple conditions in a single if statement by using the 'or' or 'and' keywords.

````
if 5 > 4 or 2 > 3:
  print('I just need one condition to be true to work!')
  
if 5 > 4 and 2 > 3:
  print('I need both conditions to be true to work!')
````

**Else and Elif** <br>
If the condition in the 'if' statement is false, the program will execute the code inside the else block.

````
if 'John' == 'John' and 3 > 5:
  print('Expression evaluated as true')
else:
  print('Expression evaluated as false')
````

Use 'elif' instead of typing 'else if'.

````
age = 40

if age < 18:
  print('Kid')
elif age < 65:
  print('Adult')
else:
  print('Yoda')
````

**While loops** <br>
The code inside the while block will be executed as long as the specified condition is true.

````
age = 0

while age < 18:
  print('Still a kid. I am now', age, 'years old.')
  age = age + 1
  
print('Doing the same stuff as before but legally')
````

**Lists**<br>
Lists are used to store multiple variables. <br>
List indexing starts at zero.

````
number = 10
pi = 3.14
name = 'John'
ok = True

list = [number, pi, name, ok]

print('Complete list:', list)
print('First element of list:', list[0])
````

To get the length of a list we use the 'len' function.

````
length = len(list)
print(length)
````

**For loops**<br>
We use for loops to iterate through each member of a list.

````
for element in list:
  print(element)
````

**Functions** <br>
We use functions to run blocks without rewriting the code and cluttering our program.

````
def shout():
  print('Bucharest School of AI rocks!')

shout()
````

````
def make_square(input):
  return input ** 2

print(make_square(3))
````
